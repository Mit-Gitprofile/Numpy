```python
import numpy as np

class ArraySearch:
    def __init__(self, arr):
        self.arr = arr

    def Search(self):
        print("Original array:\n", self.arr)
        value = int(input("Enter value for searching: "))
        result = np.where(self.arr == value)
        if result[0].size > 0:
            print(f"Value {value} found at indices: {result}")
        else:
            print(f"Value {value} not found in the array.")

# Example usage for 1D array:
arr_1d = np.array([10, 20, 30, 40, 50])
search_instance_1d = ArraySearch(arr_1d)
search_instance_1d.Search()

# Example usage for 2D array:
arr_2d = np.array([[10, 20, 30], [40, 50, 60]])
search_instance_2d = ArraySearch(arr_2d)
search_instance_2d.Search()

# Example usage for 3D array:
arr_3d = np.array([[[10, 20], [30, 40]], [[50, 60], [70, 80]]])
search_instance_3d = ArraySearch(arr_3d)
search_instance_3d.Search()
```
